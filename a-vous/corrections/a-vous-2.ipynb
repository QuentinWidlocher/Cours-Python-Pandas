### Exercice 1.1

Affichez le nombre de passagers qui ont voyag√©s en premi√®re classe, deuxi√®me classe et troisi√®me classe (Et v√©rifiez que le total correspond bien au total de ligne du tableau (890))

```python
import pandas as p
titanic = p.read_csv("../../demo-pandas/titanic.csv")

first_class = titanic[titanic["Pclass"] == 1].shape[0]
second_class = titanic[titanic["Pclass"] == 2].shape[0]
third_class = titanic[titanic["Pclass"] == 3].shape[0]

print(f"{first_class} passagers en premi√®re classe")
print(f"{second_class} passagers en deuxi√®me classe")
print(f"{third_class} passagers en troisi√®me classe")

first_class + second_class + third_class == 890
```

### Exercice 1.2

Affichez le nombre et le pourcentage d'hommes et de femmes √† avoir voyag√© √† bord. (Et v√©rifiez que le total correspond bien au total de ligne du tableau (890))

```python
is_male = titanic["Sex"] == "male"
is_female = titanic["Sex"] == "female"

total = titanic.shape[0]
males = titanic[is_male].shape[0]
females = titanic[is_female].shape[0]

print(f"Pourcentage d'hommes √† avoir voyag√©s √† bord : {round(males/total, 2) * 100}% ({males})")
print(f"Pourcentage de femmes √† avoir voyag√©es √† bord : {round(females/total, 2) * 100}% ({females})")

males + females == total
```

### Exercice 1.3

Affichez le pourcentage d'hommes et le pourcentage de femmes √† avoir surv√©cu et √† ne pas avoir surv√©cu. (Et v√©rifiez que le total correspond bien au total de ligne du tableau (890))

```python
has_survived = titanic["Survived"] == 1
has_not_survived = titanic["Survived"] == 0

surviving_males = titanic[is_male & has_survived].shape[0]
not_surviving_males = titanic[is_male & has_not_survived].shape[0]

surviving_females = titanic[is_female & has_survived].shape[0]
not_surviving_females = titanic[is_female & has_not_survived].shape[0]

print(f"Pourcentage d'hommes √† avoir surv√©cu : {round(surviving_males/total, 2) * 100}% ({surviving_males})")
print(f"Pourcentage d'hommes √† ne pas avoir surv√©cu : {round(not_surviving_males/total, 2) * 100}% ({not_surviving_males})")
print(f"Pourcentage de femmes √† avoir surv√©cu : {round(surviving_females/total, 2) * 100}% ({surviving_females})")
print(f"Pourcentage de femmes √† ne pas avoir surv√©cu : {round(not_surviving_females/total, 2) * 100}% ({not_surviving_females})")

surviving_males + not_surviving_males + surviving_females + not_surviving_females == total
```
